````python
country_sales = df['Country'].value_counts()

# Visualizar con un gráfico de barras
plt.figure(figsize=(12, 6))
sns.countplot(x='Country', data=df, order=country_sales.index)
plt.xticks(rotation=90)
plt.title('Distribución de Ventas por País')
plt.ylabel('Número de Transacciones')
plt.xlabel('País')
plt.show()
````

En el codigo anterior mostre el pais con mas ventas de la empresa, como puede ver por obvias razones el reino unido fue el que tuvo mas ventas pero tambien se ve que alemania, francia y irlanda
